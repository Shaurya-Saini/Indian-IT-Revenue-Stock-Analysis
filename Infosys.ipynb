# INFOSYS

```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("Largecap_rev.xlsx")
df
```

```python
result = pd.read_csv("large_cap_stock.csv")
result
```

```python
lis = ["INFY.NS","INFY_Return"]
in_st = result[lis]
in_st = in_st.copy()
in_st["INDEX"]=np.arange(0,11)
in_st
```

```python
in_re = df[df["COMPANY"]=="INFOSYS"]
in_re = in_re.copy()
in_re["INDEX"]=np.flip(np.arange(0,11))
in_re
```

```python
infosys = pd.merge(in_re,in_st)
infosys = infosys.sort_values("INDEX")
infosys = infosys.drop("INDEX",axis=1)
infosys = infosys.drop("COMPANY",axis=1)
infosys
```

```python
fig, ax1 = plt.subplots(figsize=(10, 5))

sns.lineplot(data=infosys, x="TIME PERIOD", y="REVENUE (in Cr)", ax=ax1, color="blue")
ax1.set_ylabel("Revenue (in Cr)", color="blue")
ax1.tick_params(axis="y", labelcolor="blue")
ax1.set_ylim(0, infosys["REVENUE (in Cr)"].max() * 1.1)

ax2 = ax1.twinx()
sns.lineplot(data=infosys, x="TIME PERIOD", y="INFY.NS", ax=ax2, color="red",linestyle="dashed")
ax2.set_ylabel("Stock price", color="red")
ax2.tick_params(axis="y", labelcolor="red")
ax2.set_ylim(0, infosys["INFY.NS"].max() * 1.1)

legend_labels = [
    plt.Line2D([0], [0], color="blue", lw=2, label="Revenue"),
    plt.Line2D([0], [0], color="red", lw=2, linestyle="dashed", label="Stock Price")
]
ax1.legend(handles=legend_labels, loc="upper left")

ax1.set_xlabel("Date")
plt.title("INFOSYS: Revenue vs. Stock Price")
```

```python
fig, ax1 = plt.subplots(figsize=(10, 5))

sns.lineplot(data=infosys, x="TIME PERIOD", y="REVENUE GROWTH RATE", ax=ax1, color="blue")
ax1.set_ylabel("Revenue growth rate", color="blue")
ax1.tick_params(axis="y", labelcolor="blue")

ax2 = ax1.twinx()
sns.lineplot(data=infosys, x="TIME PERIOD", y="INFY_Return", ax=ax2, color="red",linestyle="dashed")
ax2.set_ylabel("Stock return", color="red")
ax2.tick_params(axis="y", labelcolor="red")

min_y = min(infosys["REVENUE GROWTH RATE"].min(), infosys["INFY_Return"].min()) * 1.1
max_y = max(infosys["REVENUE GROWTH RATE"].max(), infosys["INFY_Return"].max()) * 1.1
ax1.set_ylim(min_y, max_y)
ax2.set_ylim(min_y, max_y)

legend_labels = [
    plt.Line2D([0], [0], color="blue", lw=2, label="Revenue growth rate"),
    plt.Line2D([0], [0], color="red", lw=2, linestyle="dashed", label="Stock Return")
]
ax1.legend(handles=legend_labels, loc="upper left")

ax1.set_xlabel("Date")
plt.title("INFOSYS: Revenue growth rate vs. Stock Return")
```

```python
sns.regplot(data=infosys, x="REVENUE (in Cr)", y="INFY.NS", scatter_kws={'alpha':0.6}, line_kws={"color": "red"})
plt.title("Revenue vs. Stock Price (Regression)")
```

```python
sns.regplot(data=infosys, x="REVENUE GROWTH RATE", y="INFY_Return", scatter_kws={'alpha':0.6}, line_kws={"color": "red"})
plt.title("Revenue Growth Rate vs. Stock Return (Regression)")
```

Over the past 10 years, there appears to be a slight correlation between revenue growth and stock price movement. While revenue has shown steady growth, stock prices exhibit significant fluctuations. However, the overall upward trend in stock prices aligns with revenue growth. 

There appears to be no significant relationship between revenue growth rate and stock return, indicating that stock performance is influenced by factors beyond revenue growth alone.

The variations in stock price can be attributed to other factors affecting the company and the industry, which will be discussed later.

```python
corr_matrix = infosys[["REVENUE (in Cr)", "INFY.NS", "REVENUE GROWTH RATE", "INFY_Return"]].corr()
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Matrix of Financial Metrics")
```

```python
corr1 = infosys["REVENUE (in Cr)"].corr(infosys["INFY.NS"])
corr2 = infosys["REVENUE GROWTH RATE"].corr(infosys["INFY_Return"])
print(f"Correlation between revenue and stock price: {corr1:.4f}")
print(f"Correlation between revenue growth rate and stock return: {corr2:.4f}")
```

The correlation coefficient of **0.8764** between **revenue and stock price** indicates a **strong positive relationship**, suggesting that as revenue increases over time, stock prices tend to follow an upward trend. However, the correlation of **-0.3235** between **revenue growth rate and stock return** suggests a **weak negative relationship**, implying that short-term stock returns are not significantly driven by revenue growth fluctuations and are likely influenced by other market factors, investor sentiment, and external economic conditions.
