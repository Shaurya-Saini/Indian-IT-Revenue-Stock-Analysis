## **INTRODUCTION**
This project investigates the relationship between **revenue growth** and **stock price movements** of major **Indian IT companies** over the **past decade**. By analyzing historical financial data and stock market trends, the study aims to determine the extent to which revenue growth influences stock performance. Additionally, the project explores the **impact of key global and industry-specific events**—such as economic recessions, the COVID-19 pandemic, and the adoption of AI technologies—on the stock prices of IT firms. Through this analysis, the project provides insights into how financial performance and external factors shape market valuation in the Indian IT sector.

## **RESEARCH QUESTIONS**:
1. How strongly does revenue growth impact stock prices of major Indian IT firms?
2. How do major events (earnings reports, global recessions, AI adoption) influence IT stock prices?

## **DATA COLLECTION**
This project focuses on **three large-cap and two mid-cap Indian IT companies**, selected for their readily available, consistent, and detailed financial reports. These companies provide reliable data for analyzing the relationship between revenue growth and stock price movements.

#### LARGE-CAP IT FIRMS (Market Cap < \\$10 Billion)
1. Tata Consultancy Services (TCS)
2. Infosys
3. Wipro

#### MID-CAP IT FIRMS (\\$2 Billion < Market Cap < \\$10 Billion)
1. Persistent Systems
2. Mphasis

The data for this analysis spans from FY2014 to FY2024, providing a decade-long perspective on revenue growth and stock price movements.

### Stock Data
- Large-Cap Companies: Stock data for TCS, Infosys, and Wipro was collected using the **yfinance library** in Python.

- Mid-Cap Companies: Stock data for Persistent Systems and Mphasis was obtained through **web scraping** using Python libraries such as requests and Beautiful Soup from the Yahoo Finance website.

### Revenue Data
Revenue data for all companies was extracted from their respective **annual reports**, ensuring accuracy and reliability for financial performance analysis.

This section delves into the relationship between revenue growth and stock price movements of the five selected Indian IT companies: TCS, Infosys, Wipro, Persistent Systems, and Mphasis. - Stock Returns: Yearly percentage changes in stock prices.

### IMPACT OF MAJOR EVENTS

#### 1. Global Economic Slowdown (2014-2016)
The 2015 global economic slowdown significantly impacted the IT industry as businesses reduced IT budgets, the Chinese economy weakened, and a strong US dollar eroded revenues for major tech firms. Large IT service providers like **TCS, Infosys, and Wipro** suffered due to their dependence on big enterprises, especially in banking and retail, which reduced outsourcing. This led to **slower revenue growth, missed earnings expectations, and stock price declines** for these companies. In contrast, **Persistent and Mphasis** were less affected as they focused on emerging technologies like cloud and AI, had a more diverse client base, and adapted quickly to market shifts.

The effects of this slowdown are evident in the graphs for TCS and Infosys, which show a sharp decline in revenue growth rates and stock returns during the 2014–2016 period. 

While Wipro was not as severely affected due to its diversified client base, which included smaller and mid-sized clients across various industries, reducing reliance on hard-hit sectors like banking. Additionally, Wipro had a stronger presence in emerging markets and focused on cost optimization and operational efficiency, which helped mitigate the impact of reduced IT budgets during the slowdown. It still experienced a slight dent in its revenue growth rate and stock return, reflecting the broader challenges faced by the industry during this time. 

<br>

#### 2. COVID 19
During the COVID-19 pandemic (2019–2021), the IT industry **initially faced a sharp decline in stock prices** due to global economic uncertainty and reduced client spending. However, the crisis accelerated digital transformation, cloud computing, and remote work adoption, leading to **a strong recovery**.

**Large-Cap Companies (TCS, Infosys, Wipro)**
TCS and Infosys rebounded quickly as businesses increased IT investments to support remote work and cloud migration. Their strong enterprise client base and digital transformation expertise helped them secure large contracts, driving revenue growth and stock price recovery.
Wipro also benefited from the digital shift but was initially slower to capitalize compared to TCS and Infosys. However, its diversified service offerings and focus on cloud and IT infrastructure allowed it to gain momentum later in the pandemic.

**Mid-Cap Companies (Persistent, Mphasis)**
Persistent Systems and Mphasis saw significant growth due to their expertise in cloud computing, AI, and automation. As companies accelerated digital initiatives, these firms, with their niche focus, became valuable partners for businesses undergoing rapid technological transformation.

The impact of the pandemic is clearly visible in the graphs, with an initial decline in stock prices in 2019, followed by a sharp upward trend from 2020 to 2021 as the IT sector adapted to the new normal and capitalized on the surge in demand for digital solutions.

## **CONCLUSION**
This project analyzed the relationship between revenue growth and stock price movements for five major Indian IT companies—TCS, Infosys, Wipro, Persistent Systems, and Mphasis—over the past decade (2014–2024). By examining historical financial data, stock market trends, and the impact of key global events, the study aimed to determine how strongly revenue growth influences stock performance and how external factors shape market valuation in the IT sector.

#### Key Findings
1. **Revenue growth is a strong driver of long-term stock performance** – A positive correlation was observed between revenue and stock prices for both large-cap and mid-cap companies, reinforcing the importance of consistent financial growth.

2. **Stock returns are influenced by multiple factors** – The correlation between revenue growth rate and stock returns was weaker and, at times, negative, suggesting that stock performance is influenced by factors beyond revenue growth, such as market sentiment and global economic conditions.

3. **Impact of Global Events**:
    - Events like the **2015 global economic slowdown, inflation and recession** fears highlighted the sensitivity of IT stocks to global economic conditions, with periods of reduced client spending and stock price declines.
    - The **COVID-19 pandemic** initially caused a sharp decline in stock prices but later accelerated digital transformation, leading to a strong recovery in the IT sector.

#### Limitations
- The analysis was limited to a selected set of companies and primarily evaluated data at the end of each financial year, which may not capture intra-year market fluctuations.
- Future studies could expand the scope by incorporating a larger dataset, additional financial metrics (e.g., profit margins, debt levels), and sector-specific trends for a more comprehensive analysis.

#### **FINAL THOUGHTS**
The Indian IT sector has demonstrated remarkable adaptability and consistency in terms of long-term growth over the past decade, navigating global challenges and capitalizing on technological advancements. By understanding the relationship between revenue growth, stock performance, and external factors, stakeholders can make informed decisions to drive growth and mitigate risks in an ever-evolving industry.
