```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("..\Revenue data\medcap_rev.xlsx")
df
```

```python
mp_rev = df[df["COMPANY"]=="Mphasis"]
mp_rev = mp_rev.copy()
mp_rev["Index"] =np.flip( np.arange(0,11))
mp_rev
```

```python
db = pd.read_csv('..\Stock data\mid_cap_stock.csv')
db
```

```python
lis = ["MPHASIS.NS",'MPHASIS_Return']
mp_st = db[lis]
mp_st = mp_st.copy()
mp_st["Index"] = np.arange(0,11)
mp_st
```

```python
mphasis = pd.merge(mp_rev,mp_st)
mphasis = mphasis.sort_values("Index")
mphasis = mphasis.drop("Index",axis=1)
mphasis = mphasis.drop("COMPANY",axis=1)
mphasis
```

```python
print("Average revenue growth rate over the decade = ",mphasis["Revenue growth rate"].mean())
print("Average stock return over the decade = ",mphasis["MPHASIS_Return"].mean())
```

```python
fig, ax1 = plt.subplots(figsize=(10, 5))

sns.lineplot(data=mphasis, x="DATE ", y="REVENUE (in Cr)", ax=ax1, color="blue")
ax1.set_ylabel("Revenue (in Cr)", color="blue")
ax1.tick_params(axis="y", labelcolor="blue")
ax1.set_ylim(0, mphasis["REVENUE (in Cr)"].max() * 1.1)

ax2 = ax1.twinx()
sns.lineplot(data=mphasis, x="DATE ", y="MPHASIS.NS", ax=ax2, color="red",linestyle="dashed")
ax2.set_ylabel("Stock price", color="red")
ax2.tick_params(axis="y", labelcolor="red")
ax2.set_ylim(0, mphasis["MPHASIS.NS"].max() * 1.1)

legend_labels = [
    plt.Line2D([0], [0], color="blue", lw=2, label="Revenue"),
    plt.Line2D([0], [0], color="red", lw=2, linestyle="dashed", label="Stock Price")
]
ax1.legend(handles=legend_labels, loc="upper left")

ax1.set_xlabel("Date")
plt.title("MPHASIS: Revenue vs. Stock Price")
plt.savefig("..\Graphs and Images\mphasis_rev_stk.png")
```

```python
fig, ax1 = plt.subplots(figsize=(10, 5))

sns.lineplot(data=mphasis, x="DATE ", y="Revenue growth rate", ax=ax1, color="blue")
ax1.set_ylabel("Revenue growth rate", color="blue")
ax1.tick_params(axis="y", labelcolor="blue")

ax2 = ax1.twinx()
sns.lineplot(data=mphasis, x="DATE ", y="MPHASIS_Return", ax=ax2, color="red",linestyle="dashed")
ax2.set_ylabel("Stock price", color="red")
ax2.tick_params(axis="y", labelcolor="red")

min_y = min(mphasis["Revenue growth rate"].min(), mphasis["MPHASIS_Return"].min()) * 1.1
max_y = max(mphasis["Revenue growth rate"].max(), mphasis["MPHASIS_Return"].max()) * 1.1
ax1.set_ylim(min_y, max_y)
ax2.set_ylim(min_y, max_y)

legend_labels = [
    plt.Line2D([0], [0], color="blue", lw=2, label="Revenue"),
    plt.Line2D([0], [0], color="red", lw=2, linestyle="dashed", label="Stock Price")
]
ax1.legend(handles=legend_labels, loc="upper left")

ax1.set_xlabel("Date")
plt.title("MPHASIS: Revenue growth rate vs. Stock Return")
plt.savefig("..\Graphs and Images\mphasis_rate_ret.png")
```

```python
sns.regplot(data=mphasis, x="REVENUE (in Cr)", y="MPHASIS.NS", scatter_kws={'alpha':0.6}, line_kws={"color": "red"})
plt.title("Revenue vs. Stock Price (Regression)")
plt.savefig("..\Graphs and Images\mphasis_rev_stk_reg.png")
```

```python
sns.regplot(data=mphasis, x="Revenue growth rate", y="MPHASIS_Return", scatter_kws={'alpha':0.6}, line_kws={"color": "red"})
plt.title("Revenue Growth Rate vs. Stock Return (Regression)")
plt.savefig("..\Graphs and Images\mphasis_rate_ret_reg.png")
```

Over a 10-year period, evaluating data at the end of each financial year, Mphasis' revenue shows a correlation with its stock price, as indicated by regression analysis and line plots. However, there are significant deviations from the overall trend. Additionally, the relationship between revenue growth and stock returns on a yearly basis appears to be minimal or insignificant.

```python
corr_matrix = mphasis[["REVENUE (in Cr)", "MPHASIS.NS", "Revenue growth rate", "MPHASIS_Return"]].corr()
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Matrix of Financial Metrics")
plt.savefig("..\Graphs and Images\mphasis_corr.png")
```

```python
corr1 = mphasis["REVENUE (in Cr)"].corr(mphasis["MPHASIS.NS"])
corr2 = mphasis["Revenue growth rate"].corr(mphasis["MPHASIS_Return"])
print(f"Correlation between revenue and stock price: {corr1:.4f}")
print(f"Correlation between revenue growth rate and stock return: {corr2:.4f}")
```

The correlation of **0.7943** between revenue and stock price indicates a strong positive relationship, suggesting that as Mphasis' revenue increases, its stock price generally follows a similar trend. However, the presence of significant deviations implies that other factors also influence stock price movements.  

On the other hand, the correlation of **0.0560** between revenue growth rate and stock return suggests little to no meaningful relationship. This indicates that short-term stock returns are not strongly driven by annual revenue growth, and other market dynamics, investor sentiment, or external factors likely play a more significant role in influencing stock performance.
