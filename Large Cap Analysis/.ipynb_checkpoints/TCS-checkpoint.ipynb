# TATA CONSULTANCY SERVICES (TCS)

```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("..\Revenue data\Largecap_rev.xlsx")
df
```

```python
result = pd.read_csv("..\Stock data\large_cap_stock.csv")
result
```

```python
lis = ["TCS.NS","TCS_Return"]
tcs_st = result[lis]
tcs_st = tcs_st.copy()
tcs_st["INDEX"]=np.arange(0,11)
tcs_st
```

```python
tcs_re = df[df["COMPANY"]=="TCS"]
tcs_re = tcs_re.copy()
tcs_re["INDEX"]=np.flip(np.arange(0,11))
tcs_re
```

```python
tcs = pd.merge(tcs_re,tcs_st)
tcs = tcs.sort_values("INDEX")
tcs = tcs.drop("INDEX",axis=1)
tcs = tcs.drop("COMPANY",axis=1)
tcs
```

```python
print("Average revenue growth rate = ",tcs["REVENUE GROWTH RATE"].mean())
print("Average stock return = ",tcs["TCS_Return"].mean())
```

```python
fig, ax1 = plt.subplots(figsize=(10, 5))

sns.lineplot(data=tcs, x="TIME PERIOD", y="REVENUE (in Cr)", ax=ax1, color="blue")
ax1.set_ylabel("Revenue (in Cr)", color="blue")
ax1.tick_params(axis="y", labelcolor="blue")
ax1.set_ylim(0, tcs["REVENUE (in Cr)"].max() * 1.1)

ax2 = ax1.twinx()
sns.lineplot(data=tcs, x="TIME PERIOD", y="TCS.NS", ax=ax2, color="red",linestyle="dashed")
ax2.set_ylabel("Stock price", color="red")
ax2.tick_params(axis="y", labelcolor="red")
ax2.set_ylim(0, tcs["TCS.NS"].max() * 1.1)

legend_labels = [
    plt.Line2D([0], [0], color="blue", lw=2, label="Revenue"),
    plt.Line2D([0], [0], color="red", lw=2, linestyle="dashed", label="Stock Price")
]
ax1.legend(handles=legend_labels, loc="upper left")

ax1.set_xlabel("Date")
plt.title("TCS: Revenue vs. Stock Price")
plt.savefig("..\Graphs and Images\\tcs_rev_stk.png")
```

```python
fig, ax1 = plt.subplots(figsize=(10, 5))

sns.lineplot(data=tcs, x="TIME PERIOD", y="REVENUE GROWTH RATE", ax=ax1, color="blue")
ax1.set_ylabel("Revenue growth rate", color="blue")
ax1.tick_params(axis="y", labelcolor="blue")

ax2 = ax1.twinx()
sns.lineplot(data=tcs, x="TIME PERIOD", y="TCS_Return", ax=ax2, color="red",linestyle="dashed")
ax2.set_ylabel("Stock return", color="red")
ax2.tick_params(axis="y", labelcolor="red")

min_y = min(tcs["REVENUE GROWTH RATE"].min(), tcs["TCS_Return"].min()) * 1.1
max_y = max(tcs["REVENUE GROWTH RATE"].max(), tcs["TCS_Return"].max()) * 1.1
ax1.set_ylim(min_y, max_y)
ax2.set_ylim(min_y, max_y)

legend_labels = [
    plt.Line2D([0], [0], color="blue", lw=2, label="Revenue growth rate"),
    plt.Line2D([0], [0], color="red", lw=2, linestyle="dashed", label="Stock Return")
]
ax1.legend(handles=legend_labels, loc="upper left")

ax1.set_xlabel("Date")
plt.title("TCS: Revenue growth rate vs. Stock Return")
plt.savefig("..\Graphs and Images\\tcs_rate_ret.png")
```

```python
sns.regplot(data=tcs, x="REVENUE (in Cr)", y="TCS.NS", scatter_kws={'alpha':0.6}, line_kws={"color": "red"})
plt.title("Revenue vs. Stock Price (Regression)")
plt.savefig("..\Graphs and Images\\tcs_rev_stk_reg.png")
```

```python
sns.regplot(data=tcs, x="REVENUE GROWTH RATE", y="TCS_Return", scatter_kws={'alpha':0.6}, line_kws={"color": "red"})
plt.title("Revenue Growth Rate vs. Stock Return (Regression)")
plt.savefig("..\Graphs and Images\\tcs_rate_ret_reg.png")
```

Over the past decade, there has been a noticeable correlation between revenue growth and stock price movement. While stock prices fluctuate, their overall upward trend aligns with revenue growth.  

However, the correlation between revenue growth rate and stock returns appears to be weak, suggesting that stock performance is influenced by additional factors beyond revenue growth alone.  

The fluctuations in stock prices can be attributed to various company-specific and industry-wide factors, which will be explored later.

```python
corr_matrix = tcs[["REVENUE (in Cr)", "TCS.NS", "REVENUE GROWTH RATE", "TCS_Return"]].corr()
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Matrix of Financial Metrics")
plt.savefig("..\Graphs and Images\\tcs_corr.png")
```

```python
corr1 = tcs["REVENUE (in Cr)"].corr(tcs["TCS.NS"])
corr2 = tcs["REVENUE GROWTH RATE"].corr(tcs["TCS_Return"])
print(f"Correlation between revenue and stock price: {corr1:.4f}")
print(f"Correlation between revenue growth rate and stock return: {corr2:.4f}")
```

The strong positive correlation (**0.9325**) between revenue and stock price suggests that as revenue increases, stock prices tend to rise as well. However, the negative correlation (**-0.3624**) between revenue growth rate and stock return indicates that short-term stock performance is influenced by factors beyond revenue growth alone. This suggests that while overall revenue trends drive long-term stock price movement, other market and company-specific factors play a significant role in determining stock returns.
